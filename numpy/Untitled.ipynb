```python
import numpy as np

arr1 = np.array([1,2,3,4,5])
print(arr1)
arr1.size

# find the data type 
print(arr1.dtype)
print(type(arr1))
```

```python
# Multi dimensional array
arr2 = np.array([(1,2,3,4),(1,2,3.5,5)],dtype=complex)
print(arr2)
print(arr2.dtype)
```

```python
# find the range in numpy
print(np.arange(10,30,10))
print(np.arange(10,30,5))
print(np.arange(10,30,1))
```
