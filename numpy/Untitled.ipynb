# Pandas


```python
import pandas as pd
import numpy as np

arr = np.array(['a','b','c','d'])
pd.Series(arr)
```

```python
data = {
    "student":['Rohan',"vishwa","Pradeep"],
    "age":[22,24,23]
}

df = pd.DataFrame(data)
sr = pd.Series(data)
print(df)
#   student  age
# 0    Rohan   22
# 1   vishwa   24
# 2  Pradeep   23

print(sr)
# student    [Rohan, vishwa, Pradeep]
# age                    [22, 24, 23]
```

```python
# finding the version of the pandas
print(pd.__version__)
```

```python
# Create the label in the series and Dataframe
df = pd.DataFrame(data,index=['a','b','c'])
print(df)

# student  age
# a    Rohan   22
# b   vishwa   24
# c  Pradeep   23
```

# Cleaning of Data

```python
import pandas as pd
data = pd.read_csv('students.csv')
print(data)
```

```python
#  removing empty cells row
df1 = data.dropna(subset=['Date'])
print(df1)
```

```python
# to fill the empty value.
df = pd.read_csv('students.csv')
print(df.fillna(True))
```

```python
# replace specific column
df.fillna({'Calories':0})
```

```python
#  Mean , median and mode
import pandas as pd
data = pd.read_csv('students.csv')

mean = data['Calories'].mean()
print(mean)
# print(data.fillna({'Calories':mean}))

mode = data['Calories'].mode()
print(mode)

median = data['Calories'].median()
print(median)
```

```python
# To remove duplicated value. It'll print True for duplicated value or false for unique value
print(data.duplicated())
```

```python
#  drop duplicated value
print(data.drop_duplicates())
```

```python
#  To find the correlation
data.drop('Calories',axis=1,inplace=True)
```

```python
data.drop('Date',axis=1,inplace=True)
print(data)
```

```python
#  axis = 0 : which applies to row
#  axis = 1: which applies to column
# inplace=True : Which modifies the existing DataFrame
# inplace=False : returns new DataFrame
```

```python
data.corr()
```
